```python
<!DOCTYPE html>
<html lang="vi">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Infographic: Hồ sơ Năng lực Điều hành - Nguyễn Viết Ngọc</title>
    <script src="https://cdn.tailwindcss.com"></script>
    <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
    <link rel="preconnect" href="https://fonts.googleapis.com">
    <link rel="preconnect" href="https://fonts.gstatic.com" crossorigin>
    <link href="https://fonts.googleapis.com/css2?family=Be+Vietnam+Pro:wght@400;500;700;900&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Be Vietnam Pro', sans-serif;
            background-color: #f8fafc;
        }
        .chart-container {
            position: relative;
            width: 100%;
            max-width: 600px;
            margin-left: auto;
            margin-right: auto;
            height: 350px;
            max-height: 400px;
        }
        @media (min-width: 768px) {
            .chart-container {
                height: 400px;
            }
        }
        .kpi-card {
            background-color: white;
            border-radius: 0.75rem;
            padding: 1.5rem;
            text-align: center;
            box-shadow: 0 4px 6px -1px rgb(0 0 0 / 0.1), 0 2px 4px -2px rgb(0 0 0 / 0.1);
            transition: transform 0.3s ease, box-shadow 0.3s ease;
            border-left: 5px solid;
        }
        .kpi-card:hover {
            transform: translateY(-5px);
            box-shadow: 0 10px 15px -3px rgb(0 0 0 / 0.1), 0 4px 6px -2px rgb(0 0 0 / 0.1);
        }
        .kpi-number {
            font-size: 3rem;
            font-weight: 900;
            line-height: 1;
        }
        .kpi-text {
            font-size: 1rem;
            color: #4b5563;
            margin-top: 0.5rem;
        }
        .timeline {
            position: relative;
            max-width: 1200px;
            margin: 0 auto;
        }
        .timeline::after {
            content: '';
            position: absolute;
            width: 6px;
            background-color: #118AB2;
            top: 0;
            bottom: 0;
            left: 50%;
            margin-left: -3px;
        }
        .timeline-container {
            padding: 10px 40px;
            position: relative;
            background-color: inherit;
            width: 50%;
        }
        .timeline-container.left {
            left: 0;
        }
        .timeline-container.right {
            left: 50%;
        }
        .timeline-container::after {
            content: '';
            position: absolute;
            width: 25px;
            height: 25px;
            right: -17px;
            background-color: white;
            border: 4px solid #FF6B6B;
            top: 15px;
            border-radius: 50%;
            z-index: 1;
        }
        .timeline-container.right::after {
            left: -16px;
        }
        .timeline-content {
            padding: 20px 30px;
            background-color: white;
            position: relative;
            border-radius: 6px;
            box-shadow: 0 4px 6px -1px rgb(0 0 0 / 0.1), 0 2px 4px -2px rgb(0 0 0 / 0.1);
        }
        @media screen and (max-width: 768px) {
            .timeline::after {
                left: 31px;
            }
            .timeline-container {
                width: 100%;
                padding-left: 70px;
                padding-right: 25px;
            }
            .timeline-container.right {
                left: 0%;
            }
            .timeline-container.left::after, .timeline-container.right::after {
                left: 15px;
            }
        }
    </style>
</head>
<body class="text-gray-800">

    <header class="bg-white shadow-md sticky top-0 z-20">
        <div class="container mx-auto px-6 py-4">
            <h1 class="text-2xl md:text-3xl font-bold text-[#073B4C]">Hồ sơ Năng lực Điều hành: Nguyễn Viết Ngọc</h1>
            <p class="text-lg text-[#118AB2]">Kiến trúc sư Trưởng của các Cỗ máy Tăng trưởng</p>
        </div>
    </header>

    <main class="container mx-auto p-4 md:p-8">

        <section id="kpi" class="my-12">
            <h2 class="text-3xl font-bold text-center mb-2 text-[#073B4C]">Dấu Ấn Thành Tích</h2>
            <p class="text-center text-gray-600 mb-8 max-w-3xl mx-auto">Những con số biết nói, minh chứng cho năng lực kiến tạo tăng trưởng đột phá và vận hành xuất sắc.</p>
            <div class="grid grid-cols-1 sm:grid-cols-2 lg:grid-cols-4 gap-6">
                <div class="kpi-card" style="border-color: #FF6B6B;">
                    <div class="kpi-number text-[#FF6B6B]">+130%</div>
                    <div class="kpi-text">Tăng trưởng doanh thu trong 6 tháng (từ 4 tỷ lên 9.2 tỷ/tháng)</div>
                </div>
                <div class="kpi-card" style="border-color: #FFD166;">
                    <div class="kpi-number text-[#FFD166]">15x</div>
                    <div class="kpi-text">Mở rộng hệ thống từ 2 lên 32 chi nhánh trong 18 tháng</div>
                </div>
                <div class="kpi-card" style="border-color: #06D6A0;">
                    <div class="kpi-number text-[#06D6A0]">&gt;1,000</div>
                    <div class="kpi-text">Đối tác chiến lược trong mạng lưới y tế (Bác sĩ, Phòng khám, Bệnh viện)</div>
                </div>
                <div class="kpi-card" style="border-color: #118AB2;">
                    <div class="kpi-number text-[#118AB2]">&gt;200</div>
                    <div class="kpi-text">Nhân sự được xây dựng, đào tạo và dẫn dắt trong đội ngũ hiệu suất cao</div>
                </div>
            </div>
        </section>

        <section id="growth" class="my-16 bg-white p-6 rounded-lg shadow-lg">
            <h2 class="text-3xl font-bold text-center mb-2 text-[#073B4C]">Cỗ Máy Tăng Trưởng: Từ Chiến Lược Đến Doanh Thu</h2>
            <p class="text-center text-gray-600 mb-8 max-w-3xl mx-auto">Biểu đồ thể hiện quá trình tăng trưởng doanh thu thần tốc, minh chứng cho khả năng chuyển hóa chiến lược kinh doanh thành kết quả tài chính vượt trội.</p>
            <div class="grid grid-cols-1 md:grid-cols-3 gap-8 items-center">
                <div class="md:col-span-2">
                    <div class="chart-container">
                        <canvas id="revenueChart"></canvas>
                    </div>
                </div>
                <div class="md:col-span-1 space-y-4">
                     <div class="kpi-card" style="border-color: #118AB2;">
                        <div class="kpi-number text-[#118AB2]">-10%</div>
                        <div class="kpi-text">Chi phí vận hành được tối ưu hóa qua cải tiến chuỗi cung ứng.</div>
                    </div>
                     <div class="kpi-card" style="border-color: #06D6A0;">
                        <div class="kpi-number text-[#06D6A0]">+20%</div>
                        <div class="kpi-text">Tỷ lệ giữ chân khách hàng trung thành sau khi triển khai giải pháp số hóa.</div>
                    </div>
                </div>
            </div>
        </section>

        <section id="skills" class="my-16">
            <h2 class="text-3xl font-bold text-center mb-2 text-[#073B4C]">Ma Trận Năng Lực Lãnh Đạo</h2>
            <p class="text-center text-gray-600 mb-8 max-w-3xl mx-auto">Sự kết hợp hài hòa giữa các năng lực cốt lõi, tạo nên một nhà lãnh đạo toàn diện, sẵn sàng đối mặt với mọi thách thức.</p>
             <div class="grid grid-cols-1 md:grid-cols-5 gap-8 items-center">
                <div class="md:col-span-3">
                    <div class="chart-container" style="max-width: 500px; height: 400px;">
                        <canvas id="skillsRadarChart"></canvas>
                    </div>
                </div>
                <div class="md:col-span-2 space-y-3">
                    <div class="bg-white p-4 rounded-lg shadow">
                        <h4 class="font-bold text-[#073B4C]">Hoạch định Chiến lược</h4>
                        <p class="text-sm text-gray-600">Phân tích thị trường, xây dựng mô hình kinh doanh (B2B, B2C, Franchise) & mở rộng thị trường.</p>
                    </div>
                    <div class="bg-white p-4 rounded-lg shadow">
                        <h4 class="font-bold text-[#073B4C]">Lãnh đạo & Phát triển Đội ngũ</h4>
                        <p class="text-sm text-gray-600">Xây dựng đội ngũ từ con số 0, đào tạo chuyên sâu, quản trị hiệu suất (KPIs).</p>
                    </div>
                    <div class="bg-white p-4 rounded-lg shadow">
                        <h4 class="font-bold text-[#073B4C]">Quản trị Vận hành & Tối ưu</h4>
                        <p class="text-sm text-gray-600">Tối ưu hóa quy trình, quản lý chuỗi cung ứng, tái cấu trúc & chuẩn hóa bộ máy.</p>
                    </div>
                    <div class="bg-white p-4 rounded-lg shadow">
                        <h4 class="font-bold text-[#073B4C]">Chuyên môn Y tế & Chuyển đổi số</h4>
                        <p class="text-sm text-gray-600">Am hiểu pháp lý, tư vấn giải pháp số (HIS, CRM), phát triển mạng lưới y khoa.</p>
                    </div>
                     <div class="bg-white p-4 rounded-lg shadow">
                        <h4 class="font-bold text-[#073B4C]">Quản lý Tài chính & Dữ liệu</h4>
                        <p class="text-sm text-gray-600">Quản lý ngân sách & P&L, phân tích ROI, báo cáo quản trị với Power BI.</p>
                    </div>
                </div>
            </div>
        </section>
        
        <section id="career" class="my-16">
            <h2 class="text-3xl font-bold text-center mb-2 text-[#073B4C]">Lộ Trình Sự Nghiệp: Hành Trình Tích Lũy Năng Lực</h2>
            <p class="text-center text-gray-600 mb-12 max-w-3xl mx-auto">Một lộ trình thăng tiến logic và có chủ đích, nơi mỗi vai trò là một mảnh ghép chiến lược, vun đắp nên một bộ kỹ năng toàn diện, chuẩn bị cho vai trò lãnh đạo đỉnh cao.</p>
            <div class="timeline">
                <div class="timeline-container right">
                    <div class="timeline-content">
                        <h3 class="font-bold text-xl text-[#073B4C]">DK GROUP <span class="text-base font-normal text-gray-500">(2024 - Hiện tại)</span></h3>
                        <p class="text-sm font-semibold text-[#118AB2]">Giám đốc Kinh doanh</p>
                        <p class="mt-2 text-gray-700">Kiến tạo và vận hành hệ sinh thái y tế toàn diện (Lab, Tế bào gốc, Phòng khám). Chịu trách nhiệm P&L và dẫn dắt chiến lược tăng trưởng.</p>
                    </div>
                </div>
                <div class="timeline-container left">
                    <div class="timeline-content">
                        <h3 class="font-bold text-xl text-[#073B4C]">BUSO Business Solutions <span class="text-base font-normal text-gray-500">(2021 - 2022)</span></h3>
                        <p class="text-sm font-semibold text-[#118AB2]">Giám đốc PTKD & Tư vấn số hóa</p>
                        <p class="mt-2 text-gray-700">Tiên phong trong lĩnh vực Health-Tech, tư vấn và triển khai các giải pháp số hóa (CRM, ERP, HIS) cho các đối tác y tế.</p>
                    </div>
                </div>
                 <div class="timeline-container right">
                    <div class="timeline-content">
                        <h3 class="font-bold text-xl text-[#073B4C]">MayCha <span class="text-base font-normal text-gray-500">(2019 - 2021)</span></h3>
                        <p class="text-sm font-semibold text-[#118AB2]">Trưởng phòng Vận hành & PT Hệ thống</p>
                        <p class="mt-2 text-gray-700">Dẫn dắt chiến lược nhân rộng hệ thống F&B từ 2 lên 32 chi nhánh, tối ưu chuỗi cung ứng và xây dựng đội ngũ hơn 200 người.</p>
                    </div>
                </div>
                <div class="timeline-container left">
                    <div class="timeline-content">
                        <h3 class="font-bold text-xl text-[#073B4C]">Kayspa & Swiss Consulting <span class="text-base font-normal text-gray-500">(2016 - 2018)</span></h3>
                        <p class="text-sm font-semibold text-[#118AB2]">Quản lý PTKD & Cố vấn Doanh nghiệp</p>
                        <p class="mt-2 text-gray-700">Chứng minh năng lực phát triển kinh doanh quốc tế (mở 4 chi nhánh tại Campuchia & Lào) và tư duy chiến lược (cố vấn cho SCG AG).</p>
                    </div>
                </div>
                 <div class="timeline-container right">
                    <div class="timeline-content">
                        <h3 class="font-bold text-xl text-[#073B4C]">HUSAVI <span class="text-base font-normal text-gray-500">(2013 - 2015)</span></h3>
                        <p class="text-sm font-semibold text-[#118AB2]">Chuyên viên PTKD (Thiết bị y tế)</p>
                        <p class="mt-2 text-gray-700">Đặt nền móng sự nghiệp trong ngành y tế, tạo tiền đề cho sự am hiểu sâu sắc về thị trường và được vinh danh "Quản lý tiêu biểu".</p>
                    </div>
                </div>
            </div>
        </section>

    </main>

    <footer class="bg-[#073B4C] text-white mt-16">
        <div class="container mx-auto p-6 text-center">
            <p class="font-bold">NGUYỄN VIẾT NGỌC</p>
            <p class="text-sm">Điện thoại: 0932 65 0111 | Email: vietngoc.cso@gmail.com</p>
            <p class="text-xs mt-2">Infographic được tạo bằng HTML, Tailwind CSS, và Chart.js. Hoàn toàn không sử dụng SVG hay Mermaid JS. </p>
        </div>
    </footer>

    <script>
        document.addEventListener('DOMContentLoaded', function() {
            const palette = {
                red: '#FF6B6B',
                yellow: '#FFD166',
                green: '#06D6A0',
                blue: '#118AB2',
                darkBlue: '#073B4C',
                lightGray: '#e5e7eb'
            };

            function wrapLabels(label, maxWidth = 16) {
                if (typeof label !== 'string' || label.length <= maxWidth) {
                    return label;
                }
                const words = label.split(' ');
                const lines = [];
                let currentLine = '';
                words.forEach(word => {
                    if ((currentLine + ' ' + word).trim().length > maxWidth) {
                        lines.push(currentLine.trim());
                        currentLine = word;
                    } else {
                        currentLine = (currentLine + ' ' + word).trim();
                    }
                });
                if (currentLine) {
                    lines.push(currentLine.trim());
                }
                return lines;
            }

            const tooltipTitleCallback = (tooltipItems) => {
                const item = tooltipItems[0];
                let label = item.chart.data.labels[item.dataIndex];
                if (Array.isArray(label)) {
                    return label.join(' ');
                } else {
                    return label;
                }
            };
            
            const defaultChartOptions = {
                responsive: true,
                maintainAspectRatio: false,
                plugins: {
                    legend: {
                        labels: {
                            font: {
                                family: "'Be Vietnam Pro', sans-serif",
                            }
                        }
                    },
                    tooltip: {
                        callbacks: {
                            title: tooltipTitleCallback
                        },
                        bodyFont: {
                           font: {
                                family: "'Be Vietnam Pro', sans-serif",
                            }
                        },
                        titleFont: {
                           font: {
                                family: "'Be Vietnam Pro', sans-serif",
                            }
                        }
                    }
                },
                scales: {
                    x: {
                        ticks: {
                            font: {
                                family: "'Be Vietnam Pro', sans-serif",
                            }
                        },
                        grid: {
                            display: false
                        }
                    },
                    y: {
                        ticks: {
                            font: {
                                family: "'Be Vietnam Pro', sans-serif",
                            }
                        },
                        grid: {
                            color: '#f3f4f6'
                        }
                    }
                }
            };

            const revenueCtx = document.getElementById('revenueChart').getContext('2d');
            new Chart(revenueCtx, {
                type: 'line',
                data: {
                    labels: ['Bắt đầu', 'Tháng 1', 'Tháng 2', 'Tháng 3', 'Tháng 4', 'Tháng 5', 'Tháng 6'],
                    datasets: [{
                        label: 'Doanh thu (tỷ VNĐ)',
                        data: [4, 4.8, 5.5, 6.5, 7.5, 8.5, 9.2],
                        borderColor: palette.red,
                        backgroundColor: 'rgba(255, 107, 107, 0.2)',
                        fill: true,
                        tension: 0.4
                    }]
                },
                options: { ...defaultChartOptions }
            });

            const skillsRadarCtx = document.getElementById('skillsRadarChart').getContext('2d');
            new Chart(skillsRadarCtx, {
                type: 'radar',
                data: {
                    labels: [
                        'Hoạch định Chiến lược',
                        wrapLabels('Lãnh đạo & Phát triển Đội ngũ'),
                        'Quản trị Vận hành',
                        wrapLabels('Chuyên môn Y tế & Chuyển đổi số'),
                        wrapLabels('Quản lý Tài chính & Dữ liệu')
                    ],
                    datasets: [{
                        label: 'Năng lực Cốt lõi',
                        data: [9, 9.5, 8.5, 9, 8],
                        backgroundColor: 'rgba(17, 138, 178, 0.2)',
                        borderColor: palette.blue,
                        borderWidth: 2,
                        pointBackgroundColor: palette.blue,
                        pointBorderColor: '#fff',
                        pointHoverBackgroundColor: '#fff',
                        pointHoverBorderColor: palette.blue
                    }]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                     plugins: {
                        legend: {
                            position: 'top',
                            labels: {
                                font: {
                                   family: "'Be Vietnam Pro', sans-serif",
                                }
                            }
                        },
                        tooltip: {
                             callbacks: {
                                title: tooltipTitleCallback
                            },
                            bodyFont: {
                               font: {
                                    family: "'Be Vietnam Pro', sans-serif",
                                }
                            },
                            titleFont: {
                               font: {
                                    family: "'Be Vietnam Pro', sans-serif",
                                }
                            }
                        }
                    },
                    scales: {
                        r: {
                            angleLines: {
                                color: palette.lightGray
                            },
                            grid: {
                                color: palette.lightGray
                            },
                            pointLabels: {
                                font: {
                                    size: 11,
                                    family: "'Be Vietnam Pro', sans-serif",
                                },
                                color: palette.darkBlue
                            },
                            ticks: {
                                backdropColor: '#f8fafc',
                                color: '#6b7280'
                            },
                             min: 0,
                             max: 10
                        }
                    }
                }
            });
        });
    </script>
</body>
</html>
```
